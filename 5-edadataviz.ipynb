## Exploratory Data Analysis


```python
from js import fetch
import io

URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
resp = await fetch(URL)
dataset_part_2_csv = io.BytesIO((await resp.arrayBuffer()).to_py())
df=pd.read_csv(dataset_part_2_csv)
df.head(5)
```

```python
#flightNumber vs PayloadMass
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

## Visualisation Charts to see correllations between different factors


```python
# scatter point chart with Flight Number vs launch site
sns.scatterplot(y="LaunchSite", x="FlightNumber", hue="Class", data=df)

plt.xlabel("Flight Number", fontsize=20)
plt.ylabel("Launch Site", fontsize=20)
plt.show()
```

```python
# scatter point chart with Pay Load Mass (kg) vs launch site
sns.scatterplot(y="LaunchSite", x="PayloadMass", hue="Class", data=df)

plt.xlabel("Pay Load Mass (kg)", fontsize=20)
plt.ylabel("Launch Site", fontsize=20)
plt.show()
```

```python
# Bar chart for success rate of each orbit
success_rate = df.groupby("Orbit")["Class"].mean().sort_values(ascending=False)

success_rate.plot(kind='bar')
plt.xlabel("Orbit", fontsize=20)
plt.ylabel("Success Rate", fontsize=20)
plt.title("Success Rate by Orbit Type")
plt.show()
```

```python
# scatter point chart with FlightNumber vs Orbit
sns.scatterplot(y="Orbit", x="FlightNumber", hue="Class", data=df)

plt.xlabel("Flight Number", fontsize=16)
plt.ylabel("Orbit", fontsize=16)
plt.show()
```

```python
# scatter point chart with Payload Mass vs Orbit
sns.scatterplot(y="Orbit", x="PayloadMass", hue="Class", data=df)

plt.xlabel("Payload Mass", fontsize=16)
plt.ylabel("Orbit", fontsize=16)
plt.show()
```

```python
# function to Extract years from the date 
year=[]
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
Extract_year()
df['Date'] = year
df.head()
    
```

```python
# line chart with extracted year vs success rate

yearly_success = df.groupby("Date")["Class"].mean()

plt.plot(yearly_success.index, yearly_success.values, marker ='o')

plt.xlabel("Success Rate", fontsize = 14)
plt.ylabel("Year", fontsize = 14)
plt.title("Success Rate by Year", fontsize = 20)
plt.grid(True)
plt.show()
```

## Features Engineering


```python
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
features.head()
```

### Create dummy variables to categorical columns


```python
# OneHotEncoder
features_one_hot = pd.get_dummies(
    features,
    columns=['Orbit', 'LaunchSite', 'LandingPad', 'Serial'],
    dtype=int
)

features_one_hot.head()
```

```python
# All numeric columns to float64
features_one_hot = features_one_hot.astype('float64')
features_one_hot.dtypes
```
